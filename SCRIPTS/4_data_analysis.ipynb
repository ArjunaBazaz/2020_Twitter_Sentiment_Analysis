```python
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("debate_tweets_with_sentiment.csv")

df['created_at'] = pd.to_datetime(df['created_at'])


filtered_df_trump = df[df['about'] == 'trump']
filtered_df_biden = df[df['about'] == 'biden']

cutoff_time = "2020-10-22 23:00:00"

filtered_df_trump_before = filtered_df_trump[filtered_df_trump['created_at'] < cutoff_time]
filtered_df_trump_after = filtered_df_trump[filtered_df_trump['created_at'] >= cutoff_time]

filtered_df_biden_before = filtered_df_biden[filtered_df_biden['created_at'] < cutoff_time]
filtered_df_biden_after = filtered_df_biden[filtered_df_biden['created_at'] >= cutoff_time]

print("Trump tweets before 11 PM:", len(filtered_df_trump_before))
print("Trump tweets after 11 PM:", len(filtered_df_trump_after))
print("Biden tweets before 11 PM:", len(filtered_df_biden_before))
print("Biden tweets after 11 PM:", len(filtered_df_biden_after))

filtered_df_trump_before_list = filtered_df_trump['likes'].tolist()
filtered_df_trump_after_list = filtered_df_trump['likes'].tolist()
filtered_df_biden_before_list = filtered_df_biden['likes'].tolist()
filtered_df_biden_after_list = filtered_df_biden['likes'].tolist()

#extracted sentiment scores
filtered_df_trump_before_sentiment = filtered_df_trump_before['Compound_Score'].astype(float).tolist()
filtered_df_trump_after_sentiment = filtered_df_trump_after['Compound_Score'].astype(float).tolist()
filtered_df_biden_before_sentiment = filtered_df_biden_before['Compound_Score'].astype(float).tolist()
filtered_df_biden_after_sentiment = filtered_df_biden_after['Compound_Score'].astype(float).tolist()


biden_before = filtered_df_biden_before_sentiment
biden_after = filtered_df_biden_after_sentiment
trump_before = filtered_df_trump_before_sentiment
trump_after = filtered_df_trump_after_sentiment



print(filtered_df_trump_before_list)
print(filtered_df_trump_after_list)
print(filtered_df_biden_before_list)
print(filtered_df_biden_after_list)

print("Trump sentiment before:", len(filtered_df_trump_before_sentiment))
print("Trump sentiment after:", len(filtered_df_trump_after_sentiment))
print("Biden sentiment before:", len(filtered_df_biden_before_sentiment))
print("Biden sentiment after:", len(filtered_df_biden_after_sentiment))
```

```python
# made sure lists were same length-- went by shortest
min_len_biden = min(len(filtered_df_biden_before_sentiment), len(filtered_df_biden_after_sentiment))
min_len_trump = min(len(filtered_df_trump_before_sentiment), len(filtered_df_trump_after_sentiment))

biden_before_trimmed = filtered_df_biden_before_sentiment[:min_len_biden]
biden_after_trimmed = filtered_df_biden_after_sentiment[:min_len_biden]
trump_before_trimmed = filtered_df_trump_before_sentiment[:min_len_trump]
trump_after_trimmed = filtered_df_trump_after_sentiment[:min_len_trump]

# adjusted t-test-- only if both lists have data
t_stat_biden, p_value_biden = perform_t_test(biden_before_trimmed, biden_after_trimmed) if biden_before_trimmed and biden_after_trimmed else (None, None)
t_stat_trump, p_value_trump = perform_t_test(trump_before_trimmed, trump_after_trimmed) if trump_before_trimmed and trump_after_trimmed else (None, None)

print("T-test Biden:", t_stat_biden, p_value_biden)
print("T-test Trump:", t_stat_trump, p_value_trump)


def perform_t_test(list1, list2):
  t_statistic, p_value = stats.ttest_rel(list1, list2)
  return t_statistic, p_value
```
